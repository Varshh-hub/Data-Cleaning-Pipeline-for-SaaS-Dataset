```python
class SaaSCustomer:
    def __init__(self, customer_id, name, signup_date, plan, monthly_fee, usage_hours):
        self.customer_id = customer_id
        self.name = name
        self.signup_date = signup_date
        self.plan = plan
        self.monthly_fee = monthly_fee
        self.usage_hours = usage_hours

    def validate_data(self):
        if self.name == None or self.name == "":
            self.name = "Unknown Customer"
        self.plan = self.plan.title()

        if str(self.monthly_fee).replace(".", "").isdigit():
            self.monthly_fee = float(self.monthly_fee)
        else:
            self.monthly_fee = 0.0

        if str(self.usage_hours).isdigit():
            self.usage_hours = int(self.usage_hours)
        else:
            self.usage_hours = 0

    def display_customer(self):
        print("Customer Name:", self.name)
```

```python
from datetime import datetime

def handle_missing_values(data):
    for i in range(len(data["customer_id"])):

        if data["name"][i] == None or data["name"][i] == "":
            data["name"][i] = "Unknown Customer"

        if data["usage_hours"][i] == None or data["usage_hours"][i] == "NULL":
            data["usage_hours"][i] = 0

        if str(data["monthly_fee"][i]).replace(".", "").isdigit():
            data["monthly_fee"][i] = float(data["monthly_fee"][i])
        else:
            data["monthly_fee"][i] = None

def remove_duplicates(data):
    seen = []

    i = 0
    while i < len(data["customer_id"]):

        if data["customer_id"][i] in seen:
            for key in data:
                data[key].pop(i)
        else:
            seen.append(data["customer_id"][i])
            i += 1

def standardize_dates(data):
    formats = ["%Y-%m-%d", "%Y/%m/%d", "%d-%m-%Y", "%Y%m%d"]

    for i in range(len(data["signup_date"])):

        for fmt in formats:
            try:
                date = datetime.strptime(data["signup_date"][i], fmt)
                data["signup_date"][i] = date.strftime("%Y-%m-%d")
                break
            except:
                pass

handle_missing_values(saas_data)
remove_duplicates(saas_data)
standardize_dates(saas_data)

for i in range(len(saas_data["customer_id"])):
    print(f"ID: {saas_data['customer_id'][i]} | Name: {saas_data['name'][i]} | Signup Date: {saas_data['signup_date'][i]} | Plan: {saas_data['plan'][i]} | Monthly Fee: {saas_data['monthly_fee'][i]} | Usage Hours: {saas_data['usage_hours'][i]}")
```

```python
total_customers = len(saas_data["customer_id"])
total_fee = 0
count = 0

for fee in saas_data["monthly_fee"]:
    if fee != None:
        total_fee += fee
        count += 1

average_fee = total_fee / count
total_usage = 0
for usage in saas_data["usage_hours"]:
    total_usage += int(usage)

plan_count = {}

for plan in saas_data["plan"]:
    if plan == "":
        plan = "Unknown"

    if plan in plan_count:
        plan_count[plan] += 1
    else:
        plan_count[plan] = 1

print("Total Customers:", total_customers)
print("Average Monthly Fee:", round(average_fee, 2))
print("Total Usage Hours:", total_usage)

print("\nPlan Distribution")
for plan, count in plan_count.items():
    print(plan, "-", count)
```
